```python
import csv
f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f,delimiter=',')
header = next(data)

# 최고기온
max_temp = -9999 
max_data = ''

# 최저기온
min_temp = 9999
min_data = ''

## 일교차
diff_temp = -9999
diff_data = ''

for row in data:
    if row[-1] == '': # 만약 데이터가 누락된 경우 의미없는 데이터 저장
        row[-1] = -9999
    if row[-2] == '': # 만약 데이터가 누락인 경우 의미없는 데이터 저장
        row[-2] = 9999
        
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    
    if max_temp < row[-1]:
        max_data = row[0]
        max_temp = row[-1]
    
    if min_temp > row[-2]:
        min_data = row[0]
        min_temp = row[-2]
    
    diff = row[-1] - row[-2] # 최고기온 - 최저기온
    if diff_temp < diff:
        diff_temp = diff
        diff_data = row[0]

f.close()
print('일교차가 가장 큰 날')
print(diff_data , sep='\n')

print('최저기온이 가장 낮은 날')
print(min_data , sep='\n')

print('최고기온이 가장 큰 날')
print(max_data, sep='\n')
    
        
```
